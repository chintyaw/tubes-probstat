```python
def gabung(nomor_kolom,data,opsi,del_list):
    for i in range (1, len(data.axes[0])):
        for j in range (nomor_kolom+1, nomor_kolom+opsi):
            if pd.isna(data.iloc[i,nomor_kolom]) == True:
                if pd.isna(data.iloc[i,j]) == False:
                    data.iloc[i,nomor_kolom] = data.iloc[i,j]
    for j in range (nomor_kolom+1, nomor_kolom+opsi):
        del_list.append(j)
    return data
```

```python
def ganti2(nomor_kolom,data,v,w):
    for i in range (1, len(data.axes[0])):
        if data.iloc[i,nomor_kolom]== 1:
            data.iloc[i,nomor_kolom] = v
        elif data.iloc[i,nomor_kolom] == 2:
            data.iloc[i,nomor_kolom] = w
    return data
```

# Confidence Interval

```python
def CI(kolom, variabel, n):
    kolom1 = dffinal.columns.get_loc(kolom)
    count = 0
    p = 0
    for k in range (1,n):
        if dffinal.iloc[k,kolom1] == variabel:
            count = count + 1
    p = count/n
    LCI = p - (1.96)*((p*(1-(p))/(n))*(p*(1-(p))/(n)))
    UCI = p + (1.96)*((p*(1-p)/(n)*(p*(1-(p))/(n))))
    print("Untuk n=" + str(n) + ", CI " + str(kolom)+ " "+
          str(variabel) + " ialah " + str("{:.7f}".format(LCI)) +" < p < " + str("{:.7f}".format(UCI)))
```

```python
def CIcount(kolom, count, n):
    kolom1 = dffinal.columns.get_loc(kolom)
    count = 0
    p = 0
    p = count/n
    LCI = p - (1.96)*((p*(1-(p))/(n))*(p*(1-(p))/(n)))
    UCI = p + (1.96)*((p*(1-p)/(n)*(p*(1-(p))/(n))))
    print("Untuk n=" + str(n) + ", CI " + str(kolom)+ " " + " ialah " + str("{:.7f}".format(LCI)) +" < p < " + str("{:.7f}".format(UCI)))
```
